# Handling Imbalanced Sheet

1.Up Sampling

2.Down Sampling

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(123)
samples=1000
ratio=0.8
class1=int(samples*ratio)
class2=samples-class1
```

```python
class1,class2
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
df1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=class1),
    'feature_2': np.random.normal(loc=0, scale=1, size=class1),
    'target': [0] * class1
})

df2 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=class2),
    'feature_2': np.random.normal(loc=2, scale=1, size=class2),
    'target': [1] * class2
})
```

```python
df=pd.concat([df1,df2]).reset_index(drop=True)
```

```python
df['target'].value_counts()
```

UPSAMPLING

```python
major=df[df['target']==0]
minor=df[df['target']==1]
```

```python
major
```

```python
minor
```

```python
from sklearn.utils import resample
minor_upsampled=resample(minor,replace=True, #Sample With replacement
         n_samples=len(major),
         random_state=42
        )
```

```python
minor_upsampled
```

```python
df=pd.concat([major,minor_upsampled]).reset_index(drop=True)
```

```python
print(df['target'].value_counts())
```

DOWN-SAMPLING

```python
np.random.seed(123)
samples=1000
ratio=0.8
class1=int(samples*ratio)
class2=samples-class1

## CREATE MY DATAFRAME WITH IMBALANCED DATASET
df1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=class1),
    'feature_2': np.random.normal(loc=0, scale=1, size=class1),
    'target': [0] * class1
})

df2 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=class2),
    'feature_2': np.random.normal(loc=2, scale=1, size=class2),
    'target': [1] * class2
})
df=pd.concat([df1,df2]).reset_index(drop=True)
```

```python
major=df[df['target']==0]
minor=df[df['target']==1]
```

```python
from sklearn.utils import resample
major_downsampled=resample(major,replace=False, #Sample With replacement
         n_samples=len(minor),
         random_state=42
        )
```

```python
major_downsampled
```

```python
df=pd.concat([major_downsampled,minor]).reset_index(drop=True)
```

```python
df
```

```python
print(df['target'].value_counts())
```
